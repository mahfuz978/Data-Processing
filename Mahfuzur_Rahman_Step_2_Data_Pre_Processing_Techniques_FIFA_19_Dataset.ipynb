# Release Clause Columns

```
df['Release Clause'].head()
```

```
df['Release Clause'].dtype
```

```
# The release clause column has to be converted into an integer

# define the stripping and clean up the function
def conversion(x):
  try:
    return float(x[1:-1])*100
  except:
    return np.nan
df['Release Clause'] = df['Release Clause'].apply(conversion)
```

```
print(df['Release Clause'].dtype)
print(df['Release Clause'].isnull().sum())
```

```
# Extract Relevant Data
extracted_dataset_cols = ['Crossing', 'Finishing', 'HeadingAccuracy',
                          'ShortPassing', 'Volleys', 'Dribbling',
                          'Curve', 'FKAccuracy', 'LongPassing',
                          'BallControl', 'Acceleration', 'SprintSpeed',
                          'Agility', 'Reactions', 'Balance', 'ShotPower',
                          'Jumping', 'Stamina', 'Strength', 'LongShots',
                          'Aggression', 'Interceptions', 'Positioning',
                          'Vision', 'Penalties', 'Composure', 'Marking',
                          'StandingTackle', 'SlidingTackle', 'GKDiving',
                          'GKHandling', 'GKKicking', 'GKPositioning',
                          'GKReflexes','Release Clause']


extracted_dataset = df[extracted_dataset_cols]

extracted_dataset.head()
```

```

extracted_dataset.isnull().sum()
```

```
# each column's missing values is filled with the median 
for col in extracted_dataset_cols:
    extracted_dataset[col] = extracted_dataset[col].fillna(extracted_dataset[col].median())
```

```
# check the for the number of missing values in each column
extracted_dataset.isnull().sum()
```

#Scaling the data

```
extracted_dataset.shape
```

```
# X is the set of input features
X = extracted_dataset.drop(["Release Clause"], axis=1)
X.shape
```

```
# y is the label, in this case, the Release Clause column
y = extracted_dataset['Release Clause']
y.shape
```

#Scale the input features

```
# import the MinMax Scaler from sklearn
from sklearn.preprocessing import MinMaxScaler

# initialize the min-max-scaler
scaler = MinMaxScaler()

# fit X into the scaler
scaler.fit(X)

# perform the actual scaling on X
extracted_dataset_scaled = pd.DataFrame(scaler.transform(X))

# check the preliminary stats to verify that scaling was successfully applied
extracted_dataset_scaled.describe()
```

```
extracted_dataset_scaled.head()
```

```
# columns names of scaled features
scaled_model_names = ['Crossing', 'Finishing', 'HeadingAccuracy', 
                      'ShortPassing', 'Volleys', 'Dribbling', 'Curve',
                      'FKAccuracy', 'LongPassing', 'BallControl',
                      'Acceleration', 'SprintSpeed', 'Agility', 'Reactions',
                      'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
                      'LongShots', 'Aggression', 'Interceptions', 'Positioning',
                      'Vision', 'Penalties', 'Composure', 'Marking',
                      'StandingTackle', 'SlidingTackle', 'GKDiving',
                      'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']

# assign the appropriate column names to extract
extracted_dataset_scaled.columns = scaled_model_names

# read the head of the extracted dataset after the columns have been applied
extracted_dataset_scaled.head()
```
