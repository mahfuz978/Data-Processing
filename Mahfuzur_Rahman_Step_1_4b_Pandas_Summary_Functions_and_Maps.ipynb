```
# define the function to call on the entire row
def remean_lotarea(row):
  # compute the mean of the column
  df_overallqual_mean = df.OverallQual.mean()
  
  #get the re-meaned values
  row.OverallQual = row.OverallQual-df_overallqual_mean

  #return the newly computed row
  return row
df.apply(remean_lotarea, axis='columns')
```
