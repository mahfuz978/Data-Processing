```
df1 = pd.DataFrame([['a',1],['b',2]],columns = ['letter','number'])
d2 = pd.DataFrame([['c',3],['d',4]], columns = ['letter','number'])
pd.concat([df1,d2]).reset_index(drop = True)
```

```
d1 = pd.Series([['a',1],['b',2]])
d2 = pd.Series([['c',3],['d',4]])
pd.concat([d1,d2], ignore_index=True)
```

```
d3 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']], columns=['letter', 'number', 'animal'])
  
d4 = pd.concat([df1,d3], sort = True).reset_index(drop = True)
d4.fillna('Unknown Species')
```

```
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})

df
```

```
d1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})

d1
```

```
# join `df` with `other` using appropriate suffixes on `key` column of each DataFrame
df.join(d1,lsuffix="_Primary",rsuffix="_Secondary").fillna('UnKnown')
```

```
# to join along the key column, i.e. key as the index of the joined DataFrame:
df.set_index('key').join(d1.set_index('key'))
```

```
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})

df1
```

```
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})

df2
```

```
df1.merge(df2, left_on="lkey", right_on='rkey')
```
