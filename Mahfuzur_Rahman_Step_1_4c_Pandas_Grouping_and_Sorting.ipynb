```
import pandas as pd
```

```
# read the csv from drive(google drive in this case)
df = pd.read_csv('/content/drive/MyDrive/Tech IS/Data Processing/train.csv')
df.head()
```

```
df.groupby('Neighborhood').Neighborhood.count()
```

```
df.Neighborhood.value_counts()
```

```
# Least lot area in each neighborhood
df.groupby('Neighborhood').LotArea.min()
```

```
import numpy as np
# check the first year that appears for each neighborhood
df.groupby('Neighborhood').apply(lambda x: x.YearBuilt.iloc[0])

```

```
# data to find the best conditon based on neighborhood and year built 
df.groupby(["Neighborhood", "YearBuilt"]).apply(lambda x: x.loc[x.OverallCond.idxmax()])
```

```
df.groupby("Neighborhood").SalePrice.agg([len,min,max])
```

```
# get the number of building types per year
building_types_in_the_neighborhood_per_year = df.groupby(["Neighborhood", "YearBuilt"]).BldgType.agg([len])
building_types_in_the_neighborhood_per_year.head(5)
```

```
# check the type of index of this dataset
type(building_types_in_the_neighborhood_per_year.index)
```

```
x = building_types_in_the_neighborhood_per_year.reset_index()
x
```

```
#check type of index of this data 
type(x.index)
```

```
# sort values by length
x.sort_values(by='len')
```

```
# sort values by 'len' in descending order
x.sort_values(by='len', ascending=False)
```

```
# sort by index
x.sort_index()
```

```
# sort by neighborhood and length
x.sort_values(by = ['Neighborhood', 'len'])
```
